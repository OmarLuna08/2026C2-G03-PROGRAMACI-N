```python
# Calculadora de propina inteligente

# 1. Entrada de datos
cuenta = float(input("¿Cuál es el total de la cuenta? $"))
servicio = int(input("Califica el servicio del 1 al 5: "))
# 2. Procesamiento: porcentaje según la calificación
if servicio == 5:
    porcentaje = 20
elif servicio == 4:
    porcentaje = 15
elif servicio == 3:
    porcentaje = 10
else:
    porcentaje = 5

propina = cuenta * porcentaje / 100
total_pagar = cuenta + propina

# 3. Salida de información
print("\n--- Resumen ---")
print("Porcentaje de propina:", porcentaje, "%")
print("Monto de propina: $", propina)
print("Total a pagar: $", total_pagar)


```

# Problema #4: Validador de compra con descuento

## Problema
# > $1000 aplica 20%. Entre $500-$1000 aplica 10%. Muestra precio final.

# 1. Entrada de datos
cliente = input("Ingrese el nombre del cliente: ")
monto = float(input("Ingrese el monto de la compra: $"))
frecuente = input("¿Es cliente frecuente? s/n: ").lower()

# 2. Procesamiento - calcular descuento según reglas
descuento = 0

if monto > 1000:
    descuento = monto * 0.20
elif monto >= 500:
    descuento = monto * 0.10

# Descuento extra si es cliente frecuente
if frecuente == "s":
    descuento = descuento + monto * 0.05

total_pagar = monto - descuento

# 3. Salida de información
print("\n---- FACTURA ----")
print("Cliente:", cliente)
print("Subtotal: $", monto)
print("Descuento aplicado: $", descuento)
print("Total a pagar: $", total_pagar)



```python
cliente = input("Ingrese el nombre del cliente: ")
monto = float(input("Ingrese el monto de la compra: $"))
frecuente = input("¿Es cliente frecuente? s/n: ").lower()
descuento = 0

if monto > 1000:
    descuento = monto * 0.20
elif monto >= 500:
    descuento = monto * 0.10

# Descuento extra si es cliente frecuente
if frecuente == "s":
    descuento = descuento + monto * 0.05

total_pagar = monto - descuento

# 3. Salida de información
print("\n---- FACTURA ----")
print("Cliente:", cliente)
print("Subtotal: $", monto)
print("Descuento aplicado: $", descuento)
print("Total a pagar: $", total_pagar)

```
